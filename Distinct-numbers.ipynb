```python
# Program Summary:
# This Python script includes three functions:
# 1. distinct(): Determines if all the numbers in a sequence are distinct.
# 2. dot_product(): Calculates the dot product of two arrays of integers.
# 3. norm(): Calculates the p-norm of a vector, with p=2 by default.

# Function 1: distinct()
# This function takes a sequence of numbers and checks if all elements are unique.
# It returns True if all numbers are distinct, False otherwise.
def distinct(data):
    """Return True if all elements in the sequence are distinct, else False."""
    for i in range(len(data)):
        for j in range(i + 1, len(data)):  # Compare element i with element j (j > i)
            if data[i] == data[j]:  # If any two elements are equal, return False
                return False
    return True  # If no duplicates are found, return True


# Function 2: dot_product()
# This function takes two arrays of integers and calculates the dot product.
#Define Function
def dot_product(a, b):
#Empty list to store the results
   list1 = []
#The variable i will take each value in the range
#range(len(a)) gives the sequence of numbers
#len(a) gives the length of the list
   for i in range(len(a)):
#Multiply numbers in the same position
#Append adds a single item to the end of an existing list
       list1.append(a[i] * b[i])
#Return new list
   return list1






  

# Function 3: norm()
# This function computes the p-norm of a vector, defaulting to p=2.

def norm(v, p=2):
#V input factor
#p euclidean norm


   sum_p = sum(abs(values)**p for values in v)         # sum( absolute values squared , for values are in v(#,#)
   return sum_p ** (1/p)                               #  sum to the square root of 2




   

# Driver Code:
if __name__ == "__main__":
    # Test cases for distinct() function
    list1 = [1, 3, 4, 5]
    list2 = [2, 6, 6, 7]
    print(f"Are the elements in {list1} distinct? {distinct(list1)}")  # Expected: True
    print(f"Are the elements in {list2} distinct? {distinct(list2)}")  # Expected: False

#Driver Code
a = [2, 3, 6]
b = [1, 4, 5]
print(dot_product(a, b))
  
#Driver Code
vector1 = (3, 4)
print(norm(vector1))


vector2 = (5, 12)
print(norm(vector2))
```
